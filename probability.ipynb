```python
import random

# Part 1a: Tossing a coin 10,000 times
def coin_toss_simulation(trials=10000):
    heads = 0
    tails = 0
    for _ in range(trials):
        toss = random.choice(["H", "T"])
        if toss == "H":
            heads += 1
        else:
            tails += 1
    print(f"Probability of Heads: {heads / trials}")
    print(f"Probability of Tails: {tails / trials}")

# Part 1b: Rolling two dice and getting a sum of 7
def dice_sum_simulation(trials=10000):
    success = 0
    for _ in range(trials):
        die1 = random.randint(1, 6)
        die2 = random.randint(1, 6)
        if die1 + die2 == 7:
            success += 1
    print(f"Probability of getting sum 7: {success / trials}")

coin_toss_simulation()
dice_sum_simulation()
```

```python
def probability_of_six(trials=10000):
    success = 0
    for _ in range(trials):
        rolls = [random.randint(1, 6) for _ in range(10)]
        if 6 in rolls:
            success += 1
    print(f"Probability of getting at least one 6 in 10 rolls: {success / trials}")

probability_of_six()
```

```python
def bayes_simulation(trials=1000):
    colors = ["R"] * 5 + ["G"] * 7 + ["B"] * 8
    prev_color = None
    count_blue_then_red = 0
    count_blue = 0

    for _ in range(trials):
        color = random.choice(colors)
        if prev_color == "B":
            count_blue += 1
            if color == "R":
                count_blue_then_red += 1
        prev_color = color

    if count_blue > 0:
        conditional_prob = count_blue_then_red / count_blue
        print(f"P(Red | Blue before): {conditional_prob}")
    else:
        print("No blue balls drawn as previous.")
    
bayes_simulation()
```

```python
import numpy as np

def discrete_random_variable_stats():
    values = [1, 2, 3]
    probs = [0.25, 0.35, 0.4]
    sample = np.random.choice(values, size=1000, p=probs)

    mean = np.mean(sample)
    var = np.var(sample)
    std = np.std(sample)

    print(f"Mean: {mean}")
    print(f"Variance: {var}")
    print(f"Standard Deviation: {std}")

discrete_random_variable_stats()
```

```python
import matplotlib.pyplot as plt
from scipy.stats import expon

def exponential_distribution_plot():
    data = np.random.exponential(scale=5, size=2000)

    # Histogram
    plt.hist(data, bins=50, density=True, alpha=0.6, color='skyblue', label="Histogram")

    # PDF Overlay
    x = np.linspace(0, 30, 1000)
    pdf = expon.pdf(x, scale=5)
    plt.plot(x, pdf, 'r-', label="PDF (Exponential)")

    plt.title("Exponential Distribution (mean=5)")
    plt.xlabel("Value")
    plt.ylabel("Density")
    plt.legend()
    plt.grid(True)
    plt.show()

exponential_distribution_plot()
```

```python
def central_limit_theorem_simulation():
    population = np.random.uniform(0, 1, 10000)

    # Plot original distribution
    plt.hist(population, bins=50, alpha=0.6, label="Original Distribution (Uniform)")

    sample_means = []

    for _ in range(1000):
        sample = np.random.choice(population, size=30)
        sample_means.append(np.mean(sample))

    # Plot sample mean distribution
    plt.hist(sample_means, bins=50, alpha=0.7, color='orange', label="Sample Means Distribution")

    plt.title("Central Limit Theorem Simulation")
    plt.xlabel("Value")
    plt.ylabel("Frequency")
    plt.legend()
    plt.grid(True)
    plt.show()

central_limit_theorem_simulation()
```
